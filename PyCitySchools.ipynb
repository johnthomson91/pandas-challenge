```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data_complete["school_name"].unique())
total_students = school_data_complete["student_name"].count()
total_budget = sum(school_data_complete["budget"].unique())

avg_reading_score = school_data_complete["reading_score"].mean()
passing_math = (school_data_complete["math_score"] >= 70).sum()/total_students
passing_reading = (school_data_complete["reading_score"] >= 70).sum()/total_students
overall_passing = ((school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)).sum()/total_students

district_summary_df = pd.DataFrame({
    "Total Schools":[total_schools],
    "Total Students":  '{:,}'.format(total_students),
    "Total Budget": '${:,.2f}'.format(total_budget),
    "Average Math Score": f'{avg_math_score:.2f}',
    "Average Reading Score": f'{avg_reading_score:.2f}',
    "% Passing Math":  "{0:.2%}".format(passing_math),
    "% Passing Reading": "{0:.2%}".format(passing_reading),
    "Overall Passing %": "{0:.2%}".format(overall_passing)
})
```

```python
district_summary_df
```

```python
school_summary = school_data_complete.groupby(["school_name"])

school_type = school_summary["type"].max()
school_students = school_summary["student_name"].count()
school_budget = school_summary["budget"].unique().astype(int)
per_student_budget = (school_budget/school_students).astype(int)
school_avg_math_score = school_summary["math_score"].mean()
school_avg_reading_score = school_summary["reading_score"].mean()

school_passing_math = school_data_complete[school_data_complete["math_score"] >= 70]
mathx = school_passing_math.groupby(["school_name"]).count()["student_name"]
math_percent = (mathx/school_students) * 100

school_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
readingx = school_passing_reading.groupby(["school_name"]).count()["student_name"]
reading_percent = (readingx/school_students) * 100

overall_school_passing = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
overallx = overall_school_passing .groupby(["school_name"]).count()["student_name"]
overall_percent = (overallx/school_students) * 100

school_summary_df = pd.DataFrame({
    "School Type": school_type,
    "Total Students": school_students,
    "Total School Budget": school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": school_avg_math_score,
    "Average Reading Score": school_avg_reading_score,
    "% Passing Math": math_percent,
    "% Passing Reading": reading_percent,
    "Overall Passing %": overall_percent
    
})
school_summary_df
```
